```python
import pandas as pd

df=pd.read_csv('plmap.csv')
df.head()
```

```python
import numpy as np
lati=[]
for x in df['latitude']:
    lati.append(x)
longi=[]
for x in df['longitude']:
    longi.append(x)
```

```python
lati
```

```python
longi
```

```python
import gmplot 
  
gmap3 = gmplot.GoogleMapPlotter(30.3164945, 
                                78.03219179999999, 8) 
gmap3.heatmap( lati, longi ) 
  
gmap3.draw( "map.html" )
```
