```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('fin.csv')
df.head()
```

```python
df['AQI']=df['AQI'].astype(int)
type(df['AQI'])
```

```python
def calculate_aqi(AQI):
    condition='OK'
    if(AQI<=50):
     condition='Good'
    elif(AQI>51 and AQI<100):
     condition='Moderate'
    elif(AQI>=100 and AQI<=150):
      condition='Unhealthy'
    elif(AQI>=151 and AQI<=200):
     condition='Very Unhealthy'
    else:
     condition='Hazardous'
    return condition
```

```python
df['Condition']=df['AQI'].apply(calculate_aqi)
data= df[['AQI','Condition']]
data.head()
df.head()
```
