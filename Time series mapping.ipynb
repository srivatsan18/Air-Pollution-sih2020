```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('updat.csv')
df.head()
```

```python
st=int(input("Enter the starting Year:"))
end=int(input("Enter the terminating year:"))
City=str(input("Enter the city:"))
```

```python
mydf=df[df.location==City]
mydf
```

```python
mydf=mydf[mydf.year>st]
mydf
```

```python
mydf=mydf[mydf.year<end]
mydf
```

```python
mydf.plot(kind='hist',x='year',y='AQI')
plt.show()
```
